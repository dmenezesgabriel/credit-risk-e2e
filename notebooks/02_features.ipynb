```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
s3_path = "s3://data-lake/silver/credit_risk/cleaned/"
endpoint = "http://localstack:4566"

df = pd.read_parquet(
    s3_path,
    storage_options={
        "key": "test",
        "secret": "test",
        "client_kwargs": {
            "endpoint_url": endpoint
        }
    }
)
```

```python
target = "serious_dlqin2yrs"

X = df.drop(columns=[target])
y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,  # imbalanced class
    random_state=42
)
```

```python
train = X_train.copy()
train[target] = y_train

test = X_test.copy()
test[target] = y_test
```
